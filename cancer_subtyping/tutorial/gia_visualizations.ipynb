# GIA Visualizations

This notebook runs the cancer subtyping GIA pipeline and writes outputs under `../results/gia_visualizations`. The default command is a short smoke run that can produce result files even before a checkpoint exists. For paper-style results, set `allow_untrained = False`, provide a trained checkpoint path, and increase `gia_iters`, `target_indices`, and `seeds`.

```python
from pathlib import Path
import os
import subprocess
import sys

tutorial_dir = Path.cwd()
src_dir = (tutorial_dir / '..' / 'src').resolve()
results_dir = (tutorial_dir / '..' / 'results' / 'gia_visualizations').resolve()
results_dir.mkdir(parents=True, exist_ok=True)
script = src_dir / 'run_gia_visualizations.py'
script
```

## Configure Run

```python
device = 'auto'
methods = 'keyed'
target_indices = '0'
seeds = '5001'
gia_iters = 5
gia_lr = 0.1
keyed_model_path = (tutorial_dir / '..' / 'saved_model' / 'INFL_INR_True' / 'cancer_type_fedavg.pt').resolve()
allow_untrained = True
prefix = 'gia_notebook_smoke_keydim16_keyhidden8_strength20'

cmd = [
    sys.executable, str(script),
    '--device', device,
    '--methods', methods,
    '--target_indices', target_indices,
    '--seeds', seeds,
    '--gia_iters', str(gia_iters),
    '--gia_lr', str(gia_lr),
    '--keyed_model_path', str(keyed_model_path),
    '--output_dir', str(results_dir),
    '--prefix', prefix,
]
if allow_untrained:
    cmd.append('--allow_untrained')
cmd
```

## Run GIA

```python
completed = subprocess.run(cmd, cwd=src_dir, text=True, capture_output=True)
print(completed.stdout)
print(completed.stderr)
completed.check_returncode()
```

## Inspect Outputs
