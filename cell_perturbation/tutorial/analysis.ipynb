```python
from gears import PertData, GEARS
```

```python
import gears
print(gears.__file__)
```

```python
pert_data = PertData('../data')
pert_data.load(data_name = 'adamson')
pert_data.prepare_split(split = 'simulation', seed = 1)
pert_data.get_dataloader(batch_size = 32, test_batch_size = 128)
```

```python
gears_model = GEARS(pert_data, device = 'cuda:0', 
                        weight_bias_track = False, 
                        proj_name = 'pertnet', 
                        exp_name = 'pertnet')
gears_model.model_initialize(hidden_size = 64, dp=True)                         
```

```python
gears_model.load_pretrained('../Adamson_log/Ada_INFL_8_logs_0.4', use_inr=True, use_lora=False, use_dp=False)
```

```python
gears_model.testing()
gears_model.predict([['FEV'], ['FEV', 'AHR']])
```
