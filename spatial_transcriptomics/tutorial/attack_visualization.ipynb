1. Import path and variables

```python
from pathlib import Path
import sys
from IPython.display import Image, Markdown, display

TASK_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'tutorial' else Path.cwd().resolve()
SRC_DIR = TASK_ROOT / 'src'
if str(SRC_DIR) not in sys.path:
    sys.path.append(str(SRC_DIR))

from pipeline_utils import (
    AttackVisualizationConfig,
    build_attack_visualization_command,
    run_attack_visualization,
)

print('TASK_ROOT =', TASK_ROOT)
```

2. Configure activation-FiLM INR GIA

```python
config = AttackVisualizationConfig(
    out_dir='evaluation_example_fixedsplit/gia_fed_st_demo_leaky_lenet_e5_train_sample0_key10_dpeps236_ppmleps129_i9000_progress100_fairinit_lastlinear_activation_film_inr_only',
    model='leaky_lenet',
    key_strength=10.0,
    epochs=5,
    target_split='train',
    target_index=0,
    iters=9000,
    tv_weight=3e-4,
    l2_weight=1e-6,
    grad_loss='l2+cosine',
    param_scope='last_linear',
    restarts=3,
    save_intermediate_every=100,
)
print(' '.join(build_attack_visualization_command(config)))
```

3. Run attack visualization

```python
RUN_ATTACK_VISUALIZATION = False
if RUN_ATTACK_VISUALIZATION:
    run_attack_visualization(config)
```

4. Show saved outputs

```python
out_dir = TASK_ROOT / config.out_dir
fallback_dir = Path('/deltadisk/zhangrongyu/PPML-Omics-master/Application/SpatialTranscriptomics') / config.out_dir
display_dir = out_dir if out_dir.exists() else fallback_dir
print('display_dir =', display_dir)

for name in ['metrics.md']:
    path = display_dir / name
    if path.exists():
        display(Markdown(path.read_text()))

for name in ['reconstruction_compare.png', 'attack_progress_grid.png', 'training_curve.png']:
    path = display_dir / name
    if path.exists():
        display(Markdown(f'**{name}**'))
        display(Image(filename=str(path)))
```
